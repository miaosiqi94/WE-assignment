```python
import time
import pandas as pd
import numpy as np
```

```python
pCTR = pd.read_csv('pred_dev_esm.csv')
```

```python
validationdata = pd.read_csv('validation.csv')
```

```python
averageCTR = validationdata.click.sum() / validationdata.bidid.count()
```

```python
pCTRList = list(pCTR['0'])
```

```python
def linear_bidding(pCTRList):
    bids = []
    
    basebid = 60
    
    for pCTR in pCTRList:
        bids.append(basebid * pCTR / averageCTR)

    return bids
```

```python
new = pd.DataFrame()
new['bids'] = linear_bidding(pCTRList)
validationwithbids = pd.concat([validationdata, new], axis=1)
validationwithbids
```

```python
lbs = pd.DataFrame(columns=['clicks','imps','spent','CTR','CPC','CPM'])
lists = {'clicks':[], 'imps':[], 'spent':[], 'ctr':[], 'cpc':[], 'cpm':[]}

budget = 6250000.0
imps = 0
clicks = 0
spent = 0
for i in range(0,len(validationwithbids)):
    payprice = validationwithbids.payprice[i]
    bid = validationwithbids.bids[i]
    if budget < payprice:
        break
    if bid >= payprice:
        imps = imps + 1
        clicks = clicks + validationwithbids.click[i]
        budget = budget - payprice
        spent = spent + payprice
spent = spent / 1000
ctr = ((clicks / imps) * 100).round(4).astype(str)
cpm = ((spent / imps) * 1000).round(4).astype(str)
cpc = (spent / clicks).round(4).astype(str)
lists['clicks'].append(clicks)
lists['imps'].append(imps)
lists['spent'].append(spent)
lists['ctr'].append(ctr)
lists['cpc'].append(cpc)
lists['cpm'].append(cpm)
lbs.clicks = lists['clicks']
lbs.imps = lists['imps']
lbs.spent = lists['spent']
lbs.CTR = lists['ctr']
lbs.CPC = lists['cpc']
lbs.CPM = lists['cpm']
lbs
```

```python
def ortb_bidding(pCTRList):
    bids = []
    c = 10
    l = 1e-06
    
    for pctr in pCTRList:
        bids.append(np.sqrt(((c/l)*pctr)+(c**2))-c)
    
    return bids
```

```python
new = pd.DataFrame()
new['bids'] = ortb_bidding(pCTRList)
validationwithbids = pd.concat([validationdata, new], axis=1)
validationwithbids.head()
```

```python
def ortb2_bidding(pCTRList):
    bids = []
    c = 18
    l = 8e-07
    
    for pctr in pCTRList:
        inner_eq2 = (pctr + np.sqrt(((c**2)*(l**2))+(pctr**2)))/ (c*l)
        bids.append(c*(((inner_eq2)**(1/3))-((inner_eq2)**(-1/3))))    
    return bids
```
